```python
import requests  
import csv

# user_one = "practic9"
user_one = "genbu_no_miko24"
file_name = "edges_genbu_no_miko24.csv"

url = 'https://www.livejournal.com/misc/fdata.bml?user=' + user_one
strhtml = requests.get(url)     
html_text = strhtml.text

name_list = html_text.rstrip().replace('> ','').replace('< ','').split('\n')
del name_list[0]
user1_name_list = list(set(name_list))
```

```python
def get_relations_from_user_1(user_name_list, filename):
    f = open(filename,'w')
    writer = csv.writer(f, lineterminator='\n')
    # writer.writerow(["source","target"])

    for count, user in enumerate(user_name_list):
        url = "https://www.livejournal.com/misc/fdata.bml?user=" + user
        user_text = requests.get(url).text
        user_list = user_text.rstrip().split('\n')
        del user_list[0]
        for i, record in enumerate(user_list):
            if record[0] == '>':
                record = record.replace('> ',',')
                user_list[i] = user + record
            elif record[0] == '<':
                record = record.replace('< ','')
                user_list[i] = record + ',' + user
        
        for j in user_list:
            line = j.split(",")
            writer.writerow(line)
        print("finished with user {}, {} / {}. ".format(user, count+1, len(user_name_list)))
    f.close()
```

```python
first_list = html_text.rstrip().split('\n')
del first_list[0]
for i, record in enumerate(first_list):
    if record[0] == '>':
        record = record.replace('> ',',')
        first_list[i] = user_one + record
    elif record[0] == '<':
        record = record.replace('< ','')
        first_list[i] = record + ',' + user_one
```

```python
# write the data
# finished with 7m30s

f = open('filename','w')
writer = csv.writer(f,lineterminator='\n')
writer.writerow(["source","target"])

for i in first_list:
    line = i.split(",")
    writer.writerow(line)
f.close()

get_relations_from_user_1(user1_name_list, file_name)
```

Local cluster coefficient:
$$
C_v = \frac{|\{\{x,y\}\in E:x,y\in N_v\}|}{C_{n_v}^2}
$$

```python
# calculate local clustering coefficient of 'user_one'
count = 0
with open(file_name, 'r') as f:
    reader = csv.reader(f)
    headers = next(reader)
    for row in reader:
        if row[0] in user1_name_list and row[1] in user1_name_list:
            count += 1

num_poss_edges = len(user1_name_list) * (len(user1_name_list) - 1) / 2
print("The local cluster coefficient of '{}' = {}".format(user_one ,count / num_poss_edges))
```
