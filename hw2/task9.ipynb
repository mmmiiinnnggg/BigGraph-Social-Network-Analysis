```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
```

```python
def generateGraph(t):
    G = nx.MultiDiGraph()
    G.add_node(0)
    G.add_node(1)
    G.add_edge(0, 1)
    if t == 0: return G
    num_2_degrees = [0]
    while G.number_of_nodes() < t+2:
        new_node = G.number_of_nodes()
        G.add_node(new_node)
        edge_choice = np.random.choice(G.number_of_edges())
        destination1 = list(G.edges())[edge_choice][0]
        destination2 = list(G.edges())[edge_choice][1]
        G.add_edge(new_node, destination1)
        G.add_edge(new_node, destination2)

        num_2_degree = 0
        for v in G.nodes():
            if G.degree(v) == 2:
                num_2_degree += 1
        num_2_degrees.append(num_2_degree / G.number_of_nodes())

    return G, num_2_degrees
```

```python
G1, num_2_degree_list_1 = generateGraph(1000)
G2, num_2_degree_list_2 = generateGraph(1000)
G3, num_2_degree_list_3 = generateGraph(1000)
G4, num_2_degree_list_4 = generateGraph(1000)
G5, num_2_degree_list_5 = generateGraph(1000)
plt.figure(figsize=(8,8))
plt.plot(num_2_degree_list_1)
plt.plot(num_2_degree_list_2)
plt.plot(num_2_degree_list_3)
plt.plot(num_2_degree_list_4)
plt.plot(num_2_degree_list_5)
plt.xlabel('t')
plt.ylabel('ratio: number of 2-degree nodes / number of all nodes')
plt.title("convergence of ratio: 0.5")
plt.grid()
```
